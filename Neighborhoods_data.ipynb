```python
#Choose data that we need for analysis
nbhoods=features.loc[:, ['geometry.coordinates', 'properties.name','properties.borough', 'properties.bbox']].copy()
```

```python
nbhoods
```

```python
#Rename columns
nbhoods['Borough']=nbhoods['properties.borough']
nbhoods['Neighborhood']=nbhoods['properties.name']
nbhoods['LL']=nbhoods['geometry.coordinates']
nbhoods['Bbox']=nbhoods['properties.bbox']
```

```python
nbhoods=nbhoods.drop(columns=['geometry.coordinates', 'properties.name', 'properties.borough', 'properties.bbox'])
```

```python
nbhoods['Bbox'][0]
```

```python
#Create the new columns for Lantitude and Longitude
list1 = [-73.84720052054902, 40.89470517661]
str1 = ','.join(str(e) for e in list1)
```

```python
for i in range(len(nbhoods['LL'])):
    lst_i=nbhoods['LL'][i]
    str_i=','.join(str(e) for e in lst_i)
    nbhoods['LL'][i]=str_i

                 
    
```

```python
nbhoods
```

```python
#Create columns with Longitude and latitude
# Create two lists for the loop results to be placed
lat = []
lon = []

# For each row in a varible,
for row in nbhoods['LL']:
    # Try to,
    try:
        # Split the row by comma and append
        # everything before the comma to lat
        lat.append(row.split(',')[1])
        # Split the row by comma and append
        # everything after the comma to lon
        lon.append(row.split(',')[0])
    # But if you get an error
    except:
        # append a missing value to lat
        lat.append(np.NaN)
        # append a missing value to lon
        lon.append(np.NaN)

# Create two new columns from lat and lon
nbhoods['Latitude'] = lat
nbhoods['Longitude'] = lon
```
